# Exercise 1: Predict profit of  startups

```python
import numpy as np
import pandas as pd
df_startup=pd.read_csv('D:/Megha/Desktop/DS assignments/Assignment 5 Multi Linear Regression/50_Startups.csv')
df_startup
```

```python
df_startup.isnull().sum()
```

```python
df_startup.dtypes
```

```python
df_startup[df_startup.duplicated()]
```

```python
df_startup.corr()
```

```python
import seaborn as sns
sns.pairplot(df_startup)
```

```python
df_startup
```

```python
df_startup=pd.get_dummies(df_startup,drop_first=True)
df_startup
```

```python
df_startup.rename({'R&D Spend':'RDS','Administration':'Admn','Marketing Spend':'MS','State_Florida':'ST_F','State_New York':'ST_NY'},axis=1,inplace=True)
df_startup
```

```python
import statsmodels.formula.api as smf
model=smf.ols('Profit~RDS+Admn+MS+ST_F+ST_NY',data=df_startup).fit()
```

```python
model.params
```

```python
model.pvalues
```

```python
#H0: The corresponding variable has no significant impact on profit and connot be used to make predcitions in profit
#H1: The corresponding variable has significant impact on profit and can be used to make predicitons on profit

#if pvalue> 0.05, it means that there is greater than 5% chance that the null hypothesis is true
#since pvalue <0.05, there is less thab 5% chance that the null hypothesis is true; reject H0 and accpet H1 for all variables
```

```python
model.rsquared
```

```python
df_startup['Predicted']=model.fittedvalues
df_startup['Errors']=model.resid
df_startup
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(df_startup.Profit,df_startup.Predicted,squared=False)
```

```python
df_startup.Profit.mean()
```

```python
RSME=(mean_squared_error(df_startup.Profit,df_startup.Predicted,squared=False))/df_startup.Profit.mean()
RSME
```

```python
#plot residual plt
import matplotlib.pyplot as plt
x=np.array([df_startup['Predicted']])
y=np.array([df_startup['Errors']])
plt.scatter(x,y)
plt.show()
```

```python
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q') # line = 45 to draw the diagonal line
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid>10))
```

```python
sns.boxplot(df_startup.Errors)
```

```python
sns.histplot(df_startup.Errors)
```

```python
sns.distplot(df_startup.Errors)
```

```python
model_influence=model.get_influence()
(c,_)=model_influence.cooks_distance
(c,_)
```

```python
#Plot the influencers values using stem plot
import matplotlib.pyplot as plt
x=np.arange(len(df_startup))
y=np.round(c,3)
plt.stem(x,y)
```

```python
#index and value of influencer where c is more than .5
(np.argmax(c),np.max(c))
```

```python
#since the value < 1, the model can be finalised
```

```python
from statsmodels.graphics.regressionplots import influence_plot
influence_plot(model)
plt.show()
```

```python
#predicting for new data
new_data=pd.DataFrame({'RDS': [25146,91749.16],'Admn':[126494,147852], 'MS': [658941,294919.57], 'ST_F': [0,1],'ST_NY':[0,0]},index=[3,5])
new_data
```

```python
model.predict(new_data)
```
